W tym zadaniu mieliśmy wykonać prostą funkcję, która bedzie rysowała wykresy dla poszczególnych konkretnych przedziałów.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def fun (x1,x2):    
    x=np.linspace(x1,x2)
    y=x**2 + 5
    plt.plot(x,y)
    plt.legend()
    plt.xlabel("x – argumenty")
    plt.ylabel("y – wartości")
    plt.title("wykres")
    plt.show()
przedziały=[-1,1,-6,6,0,5]
fun(przedziały[0],przedziały[1])
fun(przedziały[2],przedziały[3])
fun(przedziały[4],przedziały[5])
```
